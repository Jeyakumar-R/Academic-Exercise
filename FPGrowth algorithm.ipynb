```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
```

```python
data = [['milk', 'onion', 'apple', 'beans', 'eggs', 'banana'],
        ['dill', 'onion', 'apple', 'beans', 'eggs', 'banana'],
        ['milk', 'mango', 'beans', 'eggs'],
        ['milk', 'tomato', 'corn', 'beans', 'banana'],
        ['corn', 'onion', 'onion', 'beans', 'ice cream', 'eggs']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
re = TransactionEncoder()
re_ary = re.fit(data).transform(data)
```

```python
df = pd.DataFrame(re_ary, columns = re.columns_)
df
```

```python
fpgrowth(df, min_support = 0.6)
```

```python
frequent_items = fpgrowth(df, min_support = 0.6, use_colnames = True)
frequent_items 
```

```python
rules = association_rules(frequent_items, metric = 'confidence', min_threshold = 0.8)
rules
```

```python
from mlxtend.frequent_patterns import apriori
%timeit apriori(df, min_support = 0.6)
```
