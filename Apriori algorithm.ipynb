```python
data = [['milk', 'onion', 'apple', 'beans', 'eggs', 'banana'],
        ['dill', 'onion', 'apple', 'beans', 'eggs', 'banana'],
        ['milk', 'mango', 'beans', 'eggs'],
        ['milk', 'tomato', 'corn', 'beans', 'banana'],
        ['corn', 'onion', 'onion', 'beans', 'ice cream', 'eggs']]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
re = TransactionEncoder()
re_ary = re.fit(data).transform(data)
```

```python
re_ary
```

```python
df = pd.DataFrame(re_ary)
df
```

```python
df = pd.DataFrame(re_ary, columns = re.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
frequent_items = apriori(df, min_support = 0.6, use_colnames = True)
```

```python
frequent_items
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_items, metric = 'confidence', min_threshold = 0.8)
```
