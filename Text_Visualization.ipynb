```python
import io
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
data = pd.read_csv(io.BytesIO(uploaded['Tweets.csv']))
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
tweet = data[['airline_sentiment','text']]
```

```python
tweet.head()
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
tweet['airline_sentiment'].unique() 
```

```python
 neutral_tweet=tweet[tweet['airline_sentiment']=='neutral'] 
```

```python
positive_tweet=tweet[tweet['airline_sentiment']=='positive'] 
```

```python
 negative_tweet=tweet[tweet['airline_sentiment']=='negative'] 
```

```python
neutral = ' '.join(neutral_tweet['text'])
```

```python
 positive = ' '.join(positive_tweet['text'])
```

```python
negative = ' '.join(negative_tweet['text'])
```

```python
stopwords = set(STOPWORDS)
```

```python
 wordcloud=WordCloud(width=3000,height=1500,stopwords=stopwords).generate(positive)
 plt.figure(figsize=(15,5))
 plt.imshow(wordcloud)
 plt.axis('off')
```

```python
 wordcloud=WordCloud(width=3000,height=1500,stopwords=stopwords).generate(neutral)
 plt.figure(figsize=(15,5))
 plt.imshow(wordcloud)
 plt.axis('off')
```

```python
 wordcloud=WordCloud(width=3000,height=1500,stopwords=stopwords).generate(negative)
 plt.figure(figsize=(15,5))
 plt.imshow(wordcloud)
 plt.axis('off')
```
